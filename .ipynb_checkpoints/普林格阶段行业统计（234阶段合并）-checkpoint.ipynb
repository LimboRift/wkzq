```python
import xlwt
import xlrd
import os
import sys
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import csv
from copy import deepcopy
#####################################################
df = (pd.read_excel('行业普林格.xlsx')).fillna(0)
cn = df.columns.values.tolist()

for j in range(1,7):
    
    if j in [2,3,4]:
        p234 = []
        for i in range(len(df)):
            if df[cn[-1]][i] == j:
                p234.append(df[i:i+1:])
    else:
        exec('p'+str(j)+'=[]')
        for i in range(len(df)):
            if df[cn[-1]][i] == j:
                eval('p'+str(j)).append(df[i:i+1:])
        
def gen(df, perc):
    res = []
    temp234 = [0] * 10
    for i in range(6):
        if i not in [1,2,3]:
            temp = []
            for j in range(10):
                temp.append(sum([int(x[cn[j+1]] == i+1) for x in df]))
            res.append(temp)
        else:
            for j in range(10):
                temp234[j] += (sum([int(x[cn[j+1]] == i+1) for x in df]))
        if i == 3:
            res.append(temp234)   
    if perc == 1:
        res2 = deepcopy(res)
        for i in range(len(res)):
            for j in range(len(res[0])):
                if sum([x[j] for x in res]) != 0:
                    res2[i][j] = str(round(res[i][j]/sum([x[j] for x in res])*100))+'%'
    
        return res2
    
    return res


st = ['阶段1','阶段2、3、4','阶段5','阶段6']
    
with pd.ExcelWriter('普林格阶段行业各阶段百分比（234阶段合并）.xlsx') as writer:
    for i in range(0,6):
        if i in [1,2,3]:
            name = '阶段2、3、4'
            dff = pd.DataFrame(gen(p234,1))          #是否生成百分比，1为是
            dff.columns = cn[1:-1]
            dff.index = st
            dff.to_excel(writer, sheet_name =  '普林格阶段2、3、4')
        else:
            name = '阶段'+str(i)
            dff = pd.DataFrame(gen(eval('p'+str(i+1)),1))   
            dff.columns = cn[1:-1]
            dff.index = st
            dff.to_excel(writer, sheet_name =  '普林格阶段'+str(i+1))

writer.save()
writer.close()
```
